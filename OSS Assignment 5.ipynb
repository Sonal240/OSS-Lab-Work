```python
import numpy as np
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
import matplotlib.pyplot as plt
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

```python
# Create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize=(8, 6), dpi=80)
# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)
# Plot cosine with a blue continuous line of width 1 (pixels)
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
# Plot sine with a green continuous line of width 1 (pixels)
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")
# Set x limits
plt.xlim(-4.0, 4.0)
# Set x ticks
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
# Set y limits
plt.ylim(-1.0, 1.0)
# Set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))
# Save figure using 72 dots per inch
plt.savefig("exercise_2.png", dpi=72)
```

```python
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")
plt.legend(loc='upper left')
```

```python
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus']=False # Used to display the negative sign normally
import numpy as np
from mpl_toolkits.axisartist.axislines import SubplotZero
import numpy as np
from matplotlib.ticker import MultipleLocator, FuncFormatter

fig = plt.figure(1, (10, 6))

ax = SubplotZero(fig, 1, 1, 1)
fig.add_subplot(ax)

"""New coordinate axis"""
ax.axis["xzero"].set_visible(True)
ax.axis['yzero'].set_visible(True)

"""Coordinate Arrow"""
ax.axis["xzero"].set_axisline_style("-|>")
ax.axis["yzero"].set_axisline_style("-|>")


"""Hide axis"""

ax.axis["top",'right'].set_visible(False)

x = np.arange(-2*np.pi, 2*np.pi, 0.01)
def pi_formatter(x, pos):
    """ 
         Convert numeric value to scale text in pi/4 
    """
    m = np.round(x / (np.pi / 4))
    n = 4
    if m % 2 == 0: m, n = m / 2, n / 2
    if m % 2 == 0: m, n = m / 2, n / 2
    if m == 0:
        return "0"
    if m == 1 and n == 1:
        return "$\pi$"
    if n == 1:
        return r"$%d \pi$" % m
    if m == 1:
        return r"$\frac{\pi}{%d}$" % n
    return r"$\frac{%d \pi}{%d}$" % (m, n)


# Set the range of the two coordinate axes
plt.ylim(-3 , 3)
plt.xlim(-2*np.pi, np.max(x))

# Set the bottom margin of the graph
plt.subplots_adjust(bottom=0.15)

# The main scale is pi/4
ax.xaxis.set_major_locator(MultipleLocator(np.pi / 4))

# The main scale text is calculated with the pi_formatter function
ax.xaxis.set_major_formatter(FuncFormatter(pi_formatter))

# The minor scale is pi/20
ax.xaxis.set_minor_locator(MultipleLocator(np.pi / 20))

ax.plot(x, 1/np.sin(x),color='lightskyblue', label="$csc(x)x(-,)$")
ax.plot(x, 1/np.cos(x),color='green', label="$sec(x)x(-,)$")
ax.plot(x, np.sin(x)/np.cos(x),color='orange', label="$tan(x)x(-,)$")
ax.plot(x, np.cos(x)/np.sin(x),color='skyblue', label="$cot(x)x(-,)$")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
 
# set height of bar
Result1 = [2, 5, 8, 5]
Result2 = [3, 2, 5, 7]
 
# Set position of bar on X axis
br1 = np.arange(len(Result1))
br2 = [x + barWidth for x in br1]
 
# Make the plot
plt.bar(br1, Result1, color ='r', width = barWidth,
        edgecolor ='grey', label ='R1')
plt.bar(br2, Result2, color ='g', width = barWidth,
        edgecolor ='grey', label ='R2')

# Adding Xticks
plt.xlabel('Methods', fontweight ='bold', fontsize = 15)
plt.ylabel('Results',fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(Result1))],
        ['A', 'B', 'C', 'D'])
 
plt.legend()
plt.show()
```
